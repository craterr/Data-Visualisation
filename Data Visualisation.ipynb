# <p><center><span style="color: black; background-color: #e0bfec; padding: 22px; border: 3px solid #360747; border-radius: 5px;">Age Distribution</span></center></p>
 

```python
plt.figure(figsize = (16, 7))
plt.hist(data['Age'], bins=10,color=['#780cc4'], edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Customer Age Distribution')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```



# <p><center><span style="color: black; background-color: #e0bfec; padding: 22px; border: 2px solid #360747; border-radius: 5px;">Dividing the age into 6 groups & count each group's no. of genders</span></center></p>


```python
import pandas as pd
import matplotlib.pyplot as plt

data["Age_range"] = pd.cut(x=data["Age"], bins=[0, 20, 30, 40, 50, 60, 70], right=True)
temp = data.groupby(["Age_range", "Gender"]).size().reset_index(name="counts")
cols = temp["Age_range"].unique()
gender_count = pd.DataFrame(columns=["Age_range", "Female", "Male"])

for c in cols:
    temp_list = list(temp.loc[temp["Age_range"] == c, "counts"])
    temp_list.insert(0, str(c))
    gender_count.loc[len(gender_count)] = temp_list

x = gender_count["Age_range"]
y1 = gender_count["Female"]
y2 = gender_count["Male"]

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(x, y1, color="#d19cf7", width=0.4, label="Female", edgecolor='grey', zorder=5)
ax.bar(x, y2, color="#780cc4", width=0.4, bottom=y1, label="Male", edgecolor='grey', zorder=5)
plt.title("Gender statistics of customers by age group")
plt.grid(axis='y', alpha=0.5, ls='--')
plt.legend(frameon=False)

plt.show()
```

</span></center></p>


```python
import seaborn as sns
import matplotlib.pyplot as plt

no_of_subscribers = (data['Subscription Status'] == 'Yes').sum()
total_no_of_customers = len(data)

percentage_of_subscribers = (no_of_subscribers * 100) / total_no_of_customers
print('Subscribers percentage =', percentage_of_subscribers, '%')

subscription_counts = data['Subscription Status'].value_counts()
colors = ['#780cc4', '#d19cf7']  
subscription_counts.plot(kind='pie', autopct='%1.1f%%', colors=colors)

plt.ylabel('')
plt.title('Subscription Status')
plt.show()
```

 ### <p><span style="color: black; background-color: #c4aead ; padding: 15px; border-radius: 5px;"> Does subscription status impact purchase behavior?</span></p>


```python

palette = {"Yes": "#682F2F", "No": "#9E726F"}


plt.figure(figsize=(10, 6))
sns.boxplot(x='Subscription Status', y='Review Rating', data=data, palette=palette)
plt.title('Review Ratings by Subscription Status')
plt.show()

from scipy.stats import ttest_ind

subscribers_ratings = data[data['Subscription Status'] == 'Yes']['Review Rating']
non_subscribers_ratings = data[data['Subscription Status'] == 'No']['Review Rating']

t_stat, p_value = ttest_ind(subscribers_ratings, non_subscribers_ratings, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

from termcolor import colored, cprint
def print_color(x): return cprint(x, 'black', 'on_cyan')
 


# Interpretation of the results
if p_value < 0.05:
    print_color("There is a significant difference in review ratings between subscribers and non-subscribers.")
else:
    print_color("There is no significant difference in review ratings between subscribers and non-subscribers.")
```
